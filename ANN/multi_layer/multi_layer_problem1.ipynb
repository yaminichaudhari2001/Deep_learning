## Feature scaling

```python
def standardScalling(x):
    x_mean=np.mean(x,axis=0)
    x_std=np.std(x,axis=0)
    x_scale=(x-x_mean)/x_std
    return x_scale
```

## Train test split

```python
split_ratio=0.8
split_index=int(split_ratio*len(x))
x_train=x[:split_index].values
x_test=x[split_index:].values
y_train=y[:split_index].values
y_test=y[split_index:].values
```

## Build multi layer neural network

```python
input_size=len(x_train[0])
hidden_size1=8
hidden_size2=8
output_size=1
w1=np.random.randn(input_size,hidden_size1)
w2=np.random.randn(hidden_size1,hidden_size2)
w3=np.random.randn(hidden_size2,output_size)
```

```python
def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def forward_propogation(x_train,w1,w2,w3):
    z1=np.dot(x_train,w1)
    a1=sigmoid(z1)
    z2=np.dot(a1,w2)
    a2=sigmoid(z2)
    z3=np.dot(a2,w3)
    a3=sigmoid(z3)
    return a1,a2,a3
```

```python
def backPropogation_descent(x_train,y_train,w1,w2,w3,alpha,epoch):
    for i in range(epoch):
        a1,a2,a3=forward_propogation(x_train,w1,w2,w3)
        delta_3 =a3-y_train
        dw3=np.dot(a2.T,delta_3)
        delta_2 = np.dot(delta_3,w3.T)* sigmoid_derivative(a2)
        dw2=np.dot(a1.T,delta_2)        
        delta_1 = np.dot(delta_2, w2.T) * sigmoid_derivative(a1)
        dw1=np.dot(x_train.T,delta_1)
        w1 -= alpha * dw1
        w2 -= alpha * dw2
        w3 -= alpha * dw3
    return w1,w2,w3
```

```python
epoch=2000
alpha=0.001
w1,w2,w3=backPropogation_descent(x_train,y_train,w1,w2,w3,alpha,epoch)
```

```python
#prediction
a1,a2,a3=forward_propogation(x_test,w1,w2,w3)
y_test_pred=np.where(a3>0.5,1,0)
y_test_pred=np.array(y_test_pred)
```
